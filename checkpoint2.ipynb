```python
# remove duplicate rows
unique_df = pd.DataFrame()
for i in df['video_id'].unique():
    latest = df[df['video_id']==i].sort_values('trending_date').iloc[-1:] # only get info from most recent trending date
    unique_df = pd.concat([unique_df, latest])


print(unique_df.head())
```

```python
# we'll analyse a type of show in entertainment for our tests
unique_df['pct_likes'] = unique_df['likes'] / (unique_df['likes'] + unique_df['dislikes'])
unique_df['pct_likes']
```

```python
# group by famous shows in entertainment
channels = ['The Tonight Show Starring Jimmy Fallon'
            , 'The Late Show with Stephen Colbert'
           , 'Jimmy Kimmel Live'
           , 'Late Night with Seth Meyers']

df_late_shows = unique_df[unique_df['channelTitle'].isin(channels)].dropna()


df_late_shows.groupby('channelTitle').agg({'view_count': 'mean', 'pct_likes': 'mean', 'video_id': 'nunique'})
```

```python

import matplotlib.pyplot as plt
import scipy.stats as stats


grouped_data = df_late_shows.groupby('channelTitle')['pct_likes'].apply(list)

# Create boxplot
plt.figure(figsize=(5, 8))
plt.boxplot(grouped_data.values)
plt.xticks(range(1, len(grouped_data) + 1), grouped_data.index, rotation=90)
plt.xlabel('Channel Title')
plt.ylabel('Percentage of Likes')
plt.title('Boxplot of Percentage of Likes by Channel Title')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Anova test
"""
Compare the percent of video likes for each of the talk shows using a one-way T
ANOVA test.
"""
piv = df_late_shows.pivot_table(index='video_id', columns='channelTitle', values='pct_likes')

# One-way ANOVA
fvalue, pvalue = stats.f_oneway(piv.iloc[:,0].dropna()
                               , piv.iloc[:,1].dropna()
                               , piv.iloc[:,2].dropna()
                               , piv.iloc[:,3].dropna())
print(f'ANOVA\nTest Statistic: {fvalue:.2f}\nP-Value: {pvalue:.4f}')
# the p value is n
```
